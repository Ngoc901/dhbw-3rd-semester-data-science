```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('KC_housing_data.csv')  

dream_home = df[(df['bedrooms'] <= 2.0) & 
                (df['condition'] >= 3) & 
                (df['sqft_living'] <= 2000) & 
                (df['sqft_lot'] <= 4000) & 
                (df['floors'] == 1.0) & 
                (df['date']) & 
                ((df['yr_built'] >= 2000) | (df['yr_renovated'] >= 2000))]

print(dream_home.head())

plt.figure(figsize=(8,5))
sns.scatterplot(data=dream_home, x='sqft_living', y='price', hue='condition', palette='viridis', s=80)

plt.title('Price vs. Living Space for Recommended Homes')
plt.xlabel('Living Space (sqft)')
plt.ylabel('Price (USD)')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.seasonal import seasonal_decompose


df = pd.read_csv('KC_housing_data.csv', parse_dates=['date'])


summary = df[['price','sqft_living','bedrooms','bathrooms','condition']].describe()
print(summary)

# price per sqft
df['price_per_sqft'] = df['price'] / df['sqft_living']


df['month'] = df['date'].dt.to_period('M')
median_month = df.groupby('month')['price'].median().astype(float)


plt.figure(figsize=(10,4))
median_month.plot()
plt.title('Median Price by Month')
plt.xlabel('Month')
plt.ylabel('Median Price USD')
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()


plt.figure(figsize=(8,4))
sns.histplot(df['price'], bins=50, kde=True)
plt.title('Price Distribution')
plt.xlim(0, df['price'].quantile(0.99))  
plt.show()


corr = df[['price','sqft_living','bedrooms','bathrooms','condition','price_per_sqft']].corr()
plt.figure(figsize=(6,5))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title('Feature Correlations with Price')
plt.show()


sales_per_month = df.groupby('month').size().astype(int)

plt.figure(figsize=(10,3))
sales_per_month.plot(kind='bar')
plt.title('Number of Sales by Month')
plt.xlabel('Month')
plt.ylabel('Sales Count')
plt.tight_layout()
plt.show()

if 'days_on_market' in df.columns:
    plt.figure(figsize=(6,3))
    df.groupby(df['date'].dt.month)['days_on_market'].median().plot(marker='o')
    plt.title('Median Days on Market by Month')
    plt.ylabel('Days')
    plt.show()
```
